### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(query)
print("Number of establishments with a hygiene score of 20:", count)

# Display the first document in the results using pprint
result = db.establishments.find_one(query)
if result:
    print("First document with a hygiene score of 20:")
    pprint(result)
else:
    print("No establishments found with a hygiene score of 20.")
```

```python
# Convert the query result to a Pandas DataFrame
df = pd.DataFrame(list(db.establishments.find(query)))

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
import re

# Define the regex pattern to match variations of "London"
london_pattern = re.compile(r"London", re.IGNORECASE)

# Define the query to find establishments with London as the Local Authority and RatingValue >= 4
query = {
    "LocalAuthorityName": {"$regex": london_pattern},
    "RatingValue": {"$gte": 4}
}

# Use count_documents to get the number of documents matching the query
count = db.establishments.count_documents(query)
print("Number of establishments with London as the Local Authority and RatingValue >= 4:", count)

# Find and print the first document in the results using pprint
result = db.establishments.find_one(query)
if result:
    print("First document with London as the Local Authority and RatingValue >= 4:")
    pprint(result)
else:
    print("No establishments found with London as the Local Authority and RatingValue >= 4.")
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(db.establishments.find(query)))

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Connect to the MongoDB database
client = MongoClient("localhost", 27017)
db = client.uk_food
from bson.son import SON

# Define the latitude and longitude of "Penang Flavours"
penang_latitude = 51.490142
penang_longitude = 0.08384
degree_search = 0.01

# Define the query to search within 0.01 degree on either side of the latitude and longitude of "Penang Flavours"
query = {
    "geocode.latitude": {"$gte": penang_latitude - degree_search, "$lte": penang_latitude + degree_search},
    "geocode.longitude": {"$gte": penang_longitude - degree_search, "$lte": penang_longitude + degree_search},
    "RatingValue": 5
}

# Define the sort criteria to sort by lowest hygiene score
sort = [("scores.Hygiene", 1)]

# Execute the aggregation pipeline to sort by lowest hygiene score and limit to top 5
pipeline = [
    {"$match": query},
    {"$sort": SON(sort)},
    {"$limit": 5}
]

# Perform aggregation
results = db.establishments.aggregate(pipeline)

# Print the results
for index, establishment in enumerate(results, 1):
    print(f"Top {index} establishment:")
    pprint(establishment)
    print()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
# Define the pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
result = db.establishments.aggregate(pipeline)

# Print the number of documents in the result
total_count = sum(doc['count'] for doc in result)
print("Total number of documents:", total_count)

# Re-execute the aggregation pipeline to get the top 10 results
result = db.establishments.aggregate(pipeline)

# Print the first 10 results
print("Top 10 Local Authority areas with hygiene score of 0:")
for index, doc in enumerate(result, 1):
    print(f"{index}. {doc['_id']}: {doc['count']} establishments")
    if index == 10:
        break

```

```python
# Execute the aggregation pipeline and convert the result to a list of dictionaries
result = list(db.establishments.aggregate(pipeline))

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```
